```python
# Instead of deleting null rows or  filling the missing review rating with mean of all ratings
# we will fill the missing ratings with the median rating with respective to each category
# i.e if a clothing item review is missing we will fill it with clothing items median rating , not with the median rating of footware item
# this would be a right approach to fill the missing data.
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))

# this will fill the null ratings with their respective category median ratings

# this is how we should think in analytical mind set & solve our road blockers
```

'''
.transform(...)

This is very important.

transform() applies a function group-wise

BUT it returns a Series of the same length as the original DataFrame

The original row order and index are preserved

----------------------------------------------

lambda x: x.fillna(x.median())

This function runs once per category.

For each category:

x = all Review Rating values in that category

x.median() = median rating within that category

x.fillna(x.median()):

- Keeps existing values unchanged

- Replaces NaN values with the category median

------------------------------------------------

Ex : 

Electronics ratings: [4.5, NaN, 4.0, 5.0]
Median = 4.5
NaN ‚Üí replaced with 4.5

------------------------------------------------

Why median (data science reasoning)

Median is robust to outliers

Ratings often have skewed distributions

Better than mean when extreme ratings exist

-------------------------------------------------
Performance & best-practice note

For large datasets, this is efficient because:

Computation is vectorized

No explicit Python loops

Uses Pandas‚Äô optimized groupby engine

------------------------------------------------

one edge case is that , if all the value in the 
column are Nan then the median is also Nan.

-------------------------------------------------

'''

```python
# create a new column (age_group)

labels = ['Young Adult','Adult' , 'Middle-aged','Senior']
df['age_group'] = pd.qcut(df['age'],q=4,labels=labels)

# pd.qcut() - splits the age into 4 equal sized groups based on data distribution
# and assigns the labels we defines

# q value and the no of labes must match (imp)
```

```python
df[['age','age_group']].head(10)
```

'''
It creates a new column age_group by dividing the age column into 4 equally sized groups (quartiles) and assigns meaningful labels to each group.

This is quantile-based binning, not range-based binning.

------------------------------------------------------

pd.qcut()

qcut stands for Quantile Cut.

It splits data based on quantiles

Each bin contains approximately the same number of rows

Bin widths (age ranges) are not equal

üìå Contrast:

pd.cut() ‚Üí equal ranges
pd.qcut() ‚Üí equal counts

------------------------------------------------------

q=4

This tells Pandas:

Split the data into 4 quantiles

These are commonly called quartiles

Quantile	Description
Q1	lowest 25%
Q2	25‚Äì50%
Q3	50‚Äì75%
Q4	highest 25%

Each group contains ~25% of the data.
Each row now belongs to one of four age groups 
 Each group has roughly the same number of rows
 
------------------------------------------------------


Why use qcut instead of cut?

Prevents imbalanced groups

Useful when age distribution is skewed

Ideal for:

Segmentation

Fair comparison across groups

Modeling (avoids sparse categories)

üìä Example use cases:

Customer segmentation

Risk stratification

Behavioral analysis

-----------------------------------------------------
Common pitfalls ‚ö†Ô∏è

1Ô∏è‚É£ Duplicate values at boundaries

If many identical ages exist, you may get:

ValueError: Bin edges must be unique


‚úÖ Fix:

pd.qcut(df['age'], q=4, labels=labels, duplicates='drop')

-------------------------------------------------------

Comparison with pd.cut() :

Feature           qcut               cut

Bin type      Equal frequency    Equal width

Group size      Balanced         Can be skewed

Best for       Segmentation      Fixed ranges

Data-driven        Yes                No

--------------------------------------------------------
'''
